```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import numpy as np
import time
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Mozilla/5.0"}) # Safari/537.36. Chrome/103.0.0.0
    soup = bts(result.text, "html.parser")
    return soup
```

```python
getAndParseURL('https://www.rottentomatoes.com/browse/movies_at_home/')
```

```python
pages = ["https://www.rottentomatoes.com/browse/movies_at_home/"]
for page in range(1,100):
    pages.append("https://www.rottentomatoes.com/browse/movies_at_home/?page="+str(page))
    
pages
```

```python
from bs4 import BeautifulSoup
import requests

# URL'den HTML'i al
url = 'https://www.rottentomatoes.com/browse/movies_at_home/'
response = requests.get(url)
html = response.text

# BeautifulSoup kullanarak linki alma
soup = BeautifulSoup(html, 'html.parser')
link = soup.find('a', {'data-qa': 'discovery-media-list-item-caption'})['href']
print(link)
```

```python
product_links = []  # products yerine product_links kullanıyorum, böylece isim çakışması olmaz
for page in pages:
    html = getAndParseURL(page)
    for product_div in html.findAll("div", {"class": "js-tile-link"}):
        product_links.append("https://www.rottentomatoes.com" + product_div.a.get("href"))
        
print(product_links)
```

```python
len(product_links)
```

```python
links = []
for element in product_links:
    if element.startswith("https://www.rottentomatoes.com/"):
        links.append(element)
```

```python
from bs4 import BeautifulSoup
import requests

# Verilen URL'den HTML'i al
url = "https://www.rottentomatoes.com/m/argylle"
response = requests.get(url)
html = response.text

# BeautifulSoup kullanarak kutu ofis bilgisini alma
soup = BeautifulSoup(html, "html.parser")
label = soup.find("b", {"class": "info-item-label"}, string="Box Office (Gross USA):")
if label:
    value = label.find_next_sibling("span", {"data-qa": "movie-info-item-value"})
    if value:
        box_office = value.text.strip()
        print("Box Office (Gross USA):", box_office)
    else:
        print("Box Office bilgisi bulunamadı.")
else:
    print("Box Office etiketi bulunamadı.")
```

```python
features = []
for link in links:
    html = getAndParseURL(link)
    try:
        name = html.find("h1",{"class":"title"}).text.strip()
    except:
        name = np.nan
    try:
        genre_raw = html.find_all("span", {"class": "genre"})[0].text.strip()
        genre = ', '.join([g.strip() for g in genre_raw.split(',')])
    except:
        genre = np.nan

    try:
        revenue = html.find("span", {"data-qa": "movie-info-item-value"}, string=lambda text: text and text.startswith("$"))
        
    except:
        revenue = np.nan
    try:
        director = html.find("a",{"data-qa":"movie-info-director"}).text.strip()
    except:
        director = np.nan
    try:
        tomatometer = html.find("score-board-deprecated")['tomatometerscore'] if html.find("score-board-deprecated") else np.nan
    except:
        tomatometer = np.nan
    try:
         audience = html.find("score-board-deprecated")['audiencescore'] if html.find("score-board-deprecated") else np.nan
    except:
        audience = np.nan
        
    features.append([name, genre, revenue, director, tomatometer, audience])
    
    time.sleep(1)
    
df_columns = ["Name","Genre","Revenue","Director","Tomatometer","Audience"]

df = pd.DataFrame.from_records(features, columns=df_columns)
```
